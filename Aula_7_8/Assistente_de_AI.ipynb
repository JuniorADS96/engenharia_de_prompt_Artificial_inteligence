#**Calculadora simples**

```python
# Este bloco de código tenta executar as instruções abaixo e captura erros se algo der errado.
try:
    # Pede ao usuário para digitar o primeiro número e armazena o que ele digita como texto.
    num1_str = input("Digite o primeiro número: ")
    # Converte o texto digitado (num1_str) em um número com casas decimais (um 'float').
    num1 = float(num1_str)

    # Pede ao usuário para digitar o segundo número e armazena o que ele digita como texto.
    num2_str = input("Digite o segundo número: ")
    # Converte o texto digitado (num2_str) em um número com casas decimais (um 'float').
    num2 = float(num2_str)

    # Imprime uma linha em branco para organizar e depois exibe as opções de operação.
    print("\nEscolha a operação:")
    # Exibe a primeira opção: Soma.
    print("1 - Soma")
    # Exibe a segunda opção: Subtração.
    print("2 - Subtração")

    # Pede ao usuário para escolher entre '1' (Soma) ou '2' (Subtração).
    operacao = input("Digite '1' para soma ou '2' para subtração: ")

    # Verifica se a escolha do usuário foi '1' (para Soma).
    if operacao == '1':
        # Se for '1', calcula a soma dos dois números.
        resultado = num1 + num2
        # Imprime o resultado da soma de forma clara para o usuário.
        print(f"\nO resultado de {num1} + {num2} é: {resultado}")
    # Caso contrário, verifica se a escolha do usuário foi '2' (para Subtração).
    elif operacao == '2':
        # Se for '2', calcula a subtração do segundo número pelo primeiro.
        resultado = num1 - num2
        # Imprime o resultado da subtração de forma clara para o usuário.
        print(f"\nO resultado de {num1} - {num2} é: {resultado}")
    # Se a escolha do usuário não foi '1' nem '2'.
    else:
        # Imprime uma mensagem informando que a opção escolhida é inválida.
        print("\nOpção inválida. Por favor, escolha 1 para soma ou 2 para subtração.")
# Este bloco 'except' é ativado se houver um erro de 'ValueError', que acontece quando o usuário digita algo que não é um número.
except ValueError:
    # Imprime uma mensagem de erro específica para entrada inválida de números.
    print("\nErro: Entrada inválida. Por favor, digite apenas números.")
# Este bloco 'except' é ativado se ocorrer qualquer outro tipo de erro inesperado.
except Exception as e:
    # Imprime uma mensagem de erro genérica, mostrando qual foi o erro.
    print(f"\nOcorreu um erro inesperado: {e}")
```
