#Aprendendo sobre WHILE


```python
#Exemplo de while (Programada para parar)
num = 0 #Aqui eu defino da onde eu quero que começe contar
while num <= 10:
  print(num)
  num=num+1 #conta o numero de repetições #Aqui eu defino depois do num de quanto em quanto eu quero contar.
```

```python
#Exemplo de while (Digite zero para encerrar, e um nº aleatorio para continuar)
num = 1
while num != 0:
  num = int(input('Digite um número para continuar ou zero para encerrar:'))
```

#Aprendendo sobre FOR
range() é uma condição que eu posso definir até onde vai vai, por ex:
Quero contar até 10 coloco range(11):
Quero contar até 10 mas de 2 em 2 coloco: range (0,11,2):
Quero contar até 10 mas de 2 em 2 e que começe a partir do 4: range (4,11,2):


```python
#Exemplo de FOR
for num in range(0,20,2):#0 é: O inicio, 20 é o limite (ate onde eu quero), e 2 é de quanto em quanto eu quero contar.
  print(num)
```

```python
#Exemplo 2 de FOR
for num in range (11): #se eu não adcionar o inicio, limite e de quanto em quanto eu quero, ele vai começar no zero e contar de um em um até o nº que coloquei.
  print(num)
```
